```python
# Utilize sua própria URL se quiser
# Repositório da API: https://github.com/digitalinnovationone/santander-dev-week-2023-api
# sdw2023_api_url = 'https://sdw-2023-prd.up.railway.app'
sdw2023_api_url = 'https://jsonplaceholder.typicode.com'
```

```python
user_ids = df['UserID'].tolist()
print(user_ids)
```

```python
import requests
import json
```

https://sdw-2023-prd.up.railway.app/users/{id}

```python
def get_user(id):
  response = requests.get(f'{sdw2023_api_url}/users/{id}')
  return response.json() if response.status_code == 200 else None

users = [user for id in user_ids if (user := get_user(id)) is not None]

print(json.dumps(users, indent=2))
```

```python
import random

# Função FAKE para não gastar dinheiro nem precisar de API Key
def generate_ai_news(user):
    # Lista de frases prontas para variar um pouco
    messages=[
        f"Olá {user['name']}, invista no seu futuro hoje mesmo!",
        f"{user['name']}, investir é a melhor forma de multiplicar seu capital.",
        f"Dica especial para {user['name']}: não deixe seu dinheiro parado!"
    ]
    # Retorna uma frase aleatória
    return random.choice(messages)

for user in users:
    news = generate_ai_news(user)
    print(news)

    if 'news' not in user:
        user['news'] = []

    user['news'].append({
        "icon": "https://digitalinnovationone.github.io/santander-dev-week-2023-api/icons/credit.svg",
        "description": news
    })
```

# Load

```python
def update_user(user):
  response = requests.put(f"{sdw2023_api_url}/users/{user['id']}", json=user)
  return True if response.status_code == 200 else False

for user in users:
  success = update_user(user)
  print(f"User {user['name']} updated? {success}!")
```
